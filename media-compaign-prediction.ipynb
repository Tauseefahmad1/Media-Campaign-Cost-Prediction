```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings("ignore")
train_data = pd.read_csv('./train_dataset.csv')
print(train_data.shape)
train_data.sample(7)
```

```python
accompanied_children = (train_data['total_children']-train_data['num_children_at_home'])
train_data.insert(4, 'accompanied_children', accompanied_children)
train_data.sample(7)
```

```python
train_data.describe()
```

```python
def compare(label):
  plt.pie(train_data[label].value_counts(normalize=True).values,
        labels=train_data[label].value_counts().index.values,
        autopct="%.2f%%", radius=2)
  plt.show()
  plt.figure(figsize=(15,5))
  sns.histplot(data=train_data, x="cost", hue=label, palette="Accent", fill=False)
  ax = plt.axes()
  ax.set_facecolor('#222')
  plt.show()
```

```python
train_data.total_children.value_counts()
```

```python
compare('total_children')
```

We observe that most customers have 1 to 4 children (2 and 4 being most common) and a very small number have either 0 or 5 children.  
We also observe that for most costs, 2 and 4 total children tend to dominate the most of the costs.

```python
train_data['num_children_at_home'].value_counts(normalize=True)
```

```python
compare('num_children_at_home')
```

The feature number of children at home has 0 as the highest, weakly followed by 1 child at home. It looks like for any range of cost, 0 children at home has the major proportion contributing to all of them

```python
train_data['accompanied_children'].value_counts(normalize=True)
```

```python
compare('accompanied_children')
```

It seems like most of the cost comes from trying to lure customers with 0 children accompanying them which does make sense as those customers don't have to worry about their children nagging to go home or buy something "urgently".

```python
train_data['avg_cars_at home(approx).1'].value_counts(normalize=True)
```

Seems like most stores have minimum 1 car and some of the car owners have 4 cars.

```python
compare('avg_cars_at home(approx).1')
```

Car owners with 2 or 3 cars are always leading the cost for any range, except in rare case where 1-car owners rule the range. Perhaps, 0 and 4 cars can be viewed as an extreme case.

```python
train_data['unit_sales(in millions)'].value_counts(normalize=True)
```

```python
compare('unit_sales(in millions)')
```

It looks like 3 million units are the leading factor behind the costs. Interesting to note that 4M and 2M unit sales contribute equally or more than 3M unit sales.

```python
compare('recyclable_package')
```

Most stores have recyclable packages. Recyclable packages increases the cost of acquiring a customer.

```python
compare('low_fat')
```

If a low fat food item is not low fat, then more efforts (or more cost) is needed to get a customer. Surprisingly, there are fake low fat food items than real. Is labelling a food item as low fat and getting a customer on that basis cheaper than actually using a real low-fat food item.

```python
compare('video_store')
```

It is interesting to see there was less gaming store, hence there was higher costs to acquire customer in order to keep them interested.

```python
compare('salad_bar')
```

Most stores have salad bar and it seems having them tends to increase the costs in acquiring a customer.

```python
compare('prepared_food')
```

Most stores have prepared food which does seem to acquire customers but also are costly to have.

```python
compare('coffee_bar')
```

Most of the times, coffee bars led to increased cost of acquiring a customer, possibly as part of their maintenance, etc. In few cases, having no coffee bars leaded the cost, perhaps to retain the customers somehow.

```python
compare('florist')
```

Having flower shelves doubled the cost most of the times to acquire a customer. After a cost 105M and more, stores without flower shelves leaded the cost in those ranges, most of the time.

```python
cor = abs(train_data.corr())
plt.figure(figsize=(10,10))
sns.heatmap(cor, annot=True, fmt='.2f')
```

We observe that salar_bar and florist are highly correlated. We see few other cases of correlation, such as between coffee_bar, video_store, salad_bar, processed_food and florist. We will drop them all except coffee_bar. We will drop unit_sales as well as it shows high correlation with store_sales and slightly more correlation with other variables compared to store_sales.

```python
train_data_new = train_data.drop(['video_store', 'salad_bar', 'prepared_food', 'florist', 'unit_sales(in millions)'], axis=1)
train_data_new.sample(6)
```

```python
cor = abs(train_data_new.corr())
plt.figure(figsize=(10,10))
sns.heatmap(cor, annot=True, fmt='.2f')
```

We will finally drop total_children and accompanied_children too and begin modelling.

```python
train_data_new = train_data_new.drop(['total_children', 'accompanied_children'], axis=1)
train_data_new.sample(6)
```

# Modelling

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error as mse
from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
y = train_data_new['cost'].values
X = train_data_new.drop('cost', axis=1)
X_scaled = pd.DataFrame(RobustScaler().fit_transform(X), columns=X.columns)
X_scaled.sample(7)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.35)
```

Simple Linear

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)
y_pred = model1.predict(X_test)
mse(y_test, y_pred, squared=False)
```

Ridge

```python
params = dict(alpha=(5, 5.5, 6, 6.5, 7),
              solver=('lsqr', 'saga', 'lbfgs'))
model2 = Ridge(tol = 0.0001, max_iter=500)
gscv = GridSearchCV(model2, params, scoring='neg_mean_squared_error')
gscv.fit(X_train, y_train)
print(gscv.best_estimator_)
```

```python
model2 = gscv.best_estimator_
y_pred = model2.predict(X_test)
mse(y_test, y_pred, squared=False)
```

Lasso

```python
params = dict(alpha=(0.01, 0.05, 0.1))
model3 = Lasso(tol = 0.0001, max_iter=500, selection='random', precompute=True, warm_start=True)
gscv = GridSearchCV(model3, params, scoring='neg_mean_squared_error')
gscv.fit(X_train, y_train)
print(gscv.best_estimator_)
```

```python
model3 = gscv.best_estimator_
y_pred = model3.predict(X_test)
mse(y_test, y_pred, squared=False)
```
